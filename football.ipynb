```python
# creating a function using def for sepetating the data of "Team & Contract" column 
def teamcleaner(x):
    if x.__contains__("On Loan"):
        return int(x.split(",")[1].split()[0]),int(x.split(",")[1].split()[0])+1,x.split(",")[0][:-6]
    
    elif x.__contains__("~"):
        return int(x[-11:].replace(" ",'').split("~")[0]),int(x[-11:].replace(" ",'').split("~")[1]),x[:-11]
    else:
        return 0,0,x
```

```python
# Creating new columns for 'start, end year' and "team name" 
df[['Start Year', 'End Year', 'Team Name']] = list(map(teamcleaner,df['Team & Contract']))
```

```python
df['Joined']=pd.to_datetime(df['Joined'])
```

```python
def multiplier(x):
    if x.__contains__("M"):
        return float(x.replace("€","").replace("M",""))*1000000
    
    elif x.__contains__("K"):
        return float(x.replace("€","").replace("K",""))*1000
    else:
        return float(x.replace("€",""))
```

```python
df['Value']=list(map(multiplier,df['Value']))
```

```python
df['Wage']=list(map(multiplier,df['Wage']))
```

```python
df['Release Clause']=list(map(multiplier,df['Release Clause']))
```
